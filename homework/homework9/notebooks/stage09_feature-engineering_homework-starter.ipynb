## Average spending

```python
df['avg_spend_per_transaction'] = df['spend'] / df['transactions']
df.head()
```

### Rationale for Feature 1
Gives insight into customer behavior efficiency.
- Higher values → fewer but bigger purchases
- Lower values → frequent smaller purchases

## Spend:income ratio

```python
df['spend_income_ratio'] = df['spend'] / df['income']
df.head()
```

### Rationale for Feature 2
Captures affordability and spending tendency.
- 1 means overspending relative to income
- <1 means conservative spending

## Age based classification

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 25, 40, 60, 100],
    labels=['Youth', 'Young Adult', 'Middle-aged', 'Senior']
)
df.head()
```

### Rationale for Feature 3
- Buckets age into meaningful categories.
- Helps compare behavior across demographics

## Tax Momentum

```python
df['txn_momentum'] = (
    df.groupby('region')['transactions']
      .transform(lambda x: x.rolling(window=7, min_periods=1).mean())
)

pd.set_option('display.max_columns', None)  
pd.set_option('display.width', None)  
pd.set_option('display.colheader_justify', 'center')
(df.loc[df['region'] == 'South']).head()
```

### Rationale for Feature 4
- Captures if transactions are increasing or decreasing for each region/customer over time.
- A 7-day moving average of transactions per region
- Shows short-term momentum → useful for seasonality/trends
