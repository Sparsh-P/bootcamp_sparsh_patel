### Correlation

### CAGR, Volatility, Sharpe ratio and Max Drawdown

```python


trading_days = 252  # approx. trading days in a year

metrics = {}

for stock in pct_change.columns:
    ret = pct_change[stock].dropna()
    cum_ret = (1 + ret).prod()
    years = (ret.index[-1] - ret.index[0]).days / 365
    CAGR = cum_ret**(1/years) - 1
    vol = ret.std() * np.sqrt(trading_days)
    sharpe = CAGR / vol if vol != 0 else np.nan
    roll_max = (1 + ret).cumprod().cummax()
    dd = (1 + ret).cumprod()/roll_max - 1
    max_dd = dd.min()
    
    metrics[stock] = [CAGR, vol, sharpe, max_dd]

stock_metrics = pd.DataFrame(metrics, index=["CAGR","Volatility","Sharpe","Max Drawdown"]).T
print(stock_metrics.round(3))
```
