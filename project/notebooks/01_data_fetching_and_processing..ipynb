## API Call

```python

from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd
import os
from dotenv import load_dotenv


load_dotenv()


os.environ['KAGGLE_USERNAME'] = os.getenv('KAGGLE_USERNAME')
os.environ['KAGGLE_KEY'] = os.getenv('KAGGLE_KEY')


api = KaggleApi()
api.authenticate()


dataset_name = "stacknishant/nse-stock-historical-price-data"
download_path = "../data/raw/nse_stock_data/"



os.makedirs(download_path, exist_ok=True)


print("Downloading NSE Stock Historical Price Data...")
api.dataset_download_files(dataset_name, path=download_path, unzip=True)


csv_files = [f for f in os.listdir(download_path) if f.endswith('.csv')]
print(f"Downloaded {len(csv_files)} stock files")


print("Sample stocks available:")
for i, file in enumerate(csv_files[:10]):
    stock_name = file.replace('.csv', '')
    print(f"{i+1}. {stock_name}")


if 'RELIANCE.csv' in csv_files:
    df = pd.read_csv(os.path.join(download_path, 'RELIANCE.csv'))
    print(f"\nRELIANCE stock data preview:")
    print(f"Shape: {df.shape}")
    print(df.head())
else:
    
    if csv_files:
        sample_file = csv_files[0]
        df = pd.read_csv(os.path.join(download_path, sample_file))
        stock_name = sample_file.replace('.csv', '')
        print(f"\n{stock_name} stock data preview:")
        print(f"Shape: {df.shape}")
        print(df.head())
```

### The reason why we see NaN in almost every column is not because our data is bad. It is because a lot of these companies were not listed during the start time i.e 1991

```python
df.info(), df.describe()
```

### Here we can see a better data of all our columns
#### Our dataset contains around 1400 companies and their daily closing prices from 1991 to 2024

```python

threshold = 0.9  


missing_ratio = df.isna().mean()


cols_to_drop = missing_ratio[missing_ratio > threshold].index.tolist()


print(f"Total columns dropped: {len(cols_to_drop)}")
print("Columns dropped:", cols_to_drop)


df_emptyremoved= df.drop(columns=cols_to_drop)

print("\nRemaining columns:", df.columns.tolist())
```

### Here we see a great catch, just because the column sare empty does not mean its bad. It is just because that these IPO are relatively new and hence do not have a lot of data.


So instead, we will be going with the data from 2010 onwards

# Data filtering and cleaning

```python
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
```

```python
df_filtered = df[df['Date'] > '2009-12-31'].copy()
```

```python
df_filtered
```

### Rounding off the values to 2 decimal places for better readability

```python
# Assuming df_filtered has Date as first column
df_filtered.iloc[:, 1:] = df_filtered.iloc[:, 1:].round(2)
```

```python
df_filtered
```

```python
# Reset index to start from 1
df_filtered.reset_index(drop=True, inplace=True)  # remove old index
df_filtered.index = df_filtered.index + 1         # start from 1

# Save to CSV
df_filtered.to_csv("/Users/sparshpatel/bootcamp_sparsh_patel/project/data/processed/filtered_rounded_nse.csv", index=True)  # index will be saved
```
