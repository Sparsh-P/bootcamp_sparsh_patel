```python
import pandas as pd

# Assume df has Date as first column
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Convert all other columns to numeric
df = df.apply(pd.to_numeric, errors='coerce')

# Percent daily change
pct_change = df.pct_change().dropna()

# Daily correlation matrix
corr = pct_change.corr()

# Max Drawdown example
cummax = df.cummax()
drawdown = (df - cummax) / cummax
max_dd = drawdown.min()

print(corr.head())
print(max_dd.head())
```
